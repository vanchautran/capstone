```python
data1=data[['HouseStyle','Neighborhood','BedroomAbvGr','TotRmsAbvGrd','GrLivArea','OverallQual','YearBuilt','YearRemodAdd','SaleType','MoSold','YrSold','HeatingQC','Heating','CentralAir','Foundation','SalePrice']]
data1
```

Based onthis website for renovation the house per sqft is betwwen 20$-30$

```python
data1['Before_ren_price']=data1.SalePrice-data1.GrLivArea*20
```

```python
data1['Profit']=data1.SalePrice-data1.Before_ren_price
```

```python
data1
```

```python
data1.corr()
```

```python
f, ax = plt.subplots(figsize=(10, 5))
sns.heatmap(data1.corr(),  square=True)
```

```python
sns.set(style="white")

# Generate a mask for the upper triangle
mask = np.zeros_like(data1.corr(), dtype=np.bool)
mask[np.triu_indices_from(mask)] = True

# Set up the matplotlib figure
f, ax = plt.subplots(figsize=(15, 10))

# Generate a custom diverging colormap
cmap = sns.diverging_palette(220, 10, as_cmap=True)

# Draw the heatmap with the mask and correct aspect ratio
sns.heatmap(data1.corr(), mask=mask, cmap=cmap, vmax=.3, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5}, annot=True)

plt.show()
```
